```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# match played in every season
plt.figure(figsize = (10,8))
data = df.groupby(['match_id','season']).count().index.droplevel(level = 0).value_counts().sort_index()
sns.barplot(y=data.index, x = data , orient = 'h',color = 'green')
plt.xlabel('total match played')
plt.ylabel('IPL Season year')
plt.show()
```

```python
# match played in venue
sd=df.groupby(['venue','match_id']).count().droplevel(level=1).index.value_counts()
sd
```

```python
# matches played by each team
plt.figure(figsize= (10,8))
data = df['bowling_team'].value_counts().sort_values(ascending = False)
sns.barplot(y = data.index,x = data,orient ="h")
plt.title("Total ball bowled by Team",color = "green")
plt.xlabel("Total balls")
plt.ylabel("Team")
plt.show()
```

```python
# most run scored by team
most_run = df.groupby(['batting_team'])['run'].sum().sort_values(ascending = False)
most_run
```

```python
# most runs scored by team by chart
plt.figure(figsize=(10,10))
plt.pie(x=most_run,labels=most_run.index)
plt.legend(most_run)
plt.show()
```

```python
# match played by team
df.groupby(['bowling_team','match_id']).count().droplevel(level=1).index.value_counts()
```

```python
# Most Runs by Batsman
plt.figure(figsize= (12,8))
data = df.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = False)[:10]
sns.barplot(y = data.index, x = data , orient = 'h')
plt.title("Most Runs By Batsman In IPL Season",size = 20 , color = "violet")
plt.ylabel('batsman',size = 12,color = "green")
plt.xlabel('Runs',size = 16 , color = "green")
plt.show()
```

```python
# Average_runs_by_team

total_runs = df.groupby(['batting_team'])['run'].sum().sort_values(ascending = False)
total_matches = df.groupby(['batting_team','match_id']).count().droplevel(level=1).index.value_counts()
average_runs = total_runs / total_matches
as_order = average_runs.sort_values(ascending = False)
Average_runs = pd.DataFrame(as_order)
Average_runs
```

```python
# average runs by team
plt.figure(figsize= (10,8))
sns.barplot(y = as_order.index , x = as_order , orient = 'h')
plt.title('Average runs by team in all season',size = 18 , color = 'green')
plt.xlabel('Average runs',size = 16 , color = "green")
plt.ylabel('Teams',size = 16 , color = "green")
plt.show()
```

```python
data = df.groupby(['match_id','striker'])['runs_off_bat'].sum().sort_values(ascending = False)
data1 = pd.DataFrame(data)
g100 = data1['runs_off_bat']
g100
```

```python
# avg by team in power play
df[df['over']<6].groupby(['match_id','batting_team']).sum()['run'].groupby('batting_team').mean().sort_values(ascending = False)[2:]
```

```python
# most ipl century by a player
runs = df.groupby(['striker','match_id'])['runs_off_bat'].sum()
runs[runs>= 100].droplevel(level = 1).groupby('striker').count().sort_values(ascending = False)[:10]
```

```python
# most ipl fifty by player
plt.figure(figsize = (10,8))
runs = df.groupby(['striker','start_date'])['runs_off_bat'].sum()
Data = runs[runs >= 50].droplevel(level = 1).groupby('striker').count().sort_values(ascending = False)[:10]
sns.barplot(y = Data.index , x = Data , orient = 'h')
plt.title('most fifty by players',size = 20 , color = "green")
plt.xlabel('numbers of fifty',size = 16 , color = "green")
plt.ylabel('Player',size = 16 , color = "green")
plt.show()
```

```python
# most six by players
df[df['runs_off_bat'] == 6].groupby(['start_date','striker']).count()['season'].sort_values(ascending = False).droplevel(level = 0)[:10]
```

```python
# most fours by player
plt.figure(figsize = (10,8))
data = df[df['runs_off_bat'] == 4]['striker'].value_counts()[:10]
sns.barplot(y = data.index , x = data , orient = 'h')
plt.xlabel('numbers of four')
plt.ylabel('player name')
plt.show()
```

```python
# most out of team
plt.figure(figsize = (10,8))
lst = 'caught , bowled , lbw , stumped , caught and bowled , hit wicket'
data = df[df['wicket_type'].apply(lambda x: True if x in lst and x!= ' '  else False)]['bowling_team'].value_counts()
df.groupby(['batting_team'])['extras'].agg('sum').sort_values(ascending = False)
sns.barplot(y = data.index , x = data , orient = 'h')
plt.ylabel('Teams')
plt.xlabel('Wickets')
plt.show()
```
